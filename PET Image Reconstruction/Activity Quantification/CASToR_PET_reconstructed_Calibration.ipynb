### Calibration of PET CASToR reconstructed image using CF ans Rescale Slope

```python
import numpy as np
import nibabel as nib

# -------------------------------------------------
# USER PARAMETERS
# -------------------------------------------------

# Scanner sensitivity (cps / MBq)
sensitivity = 5080

# voxel size (mm)
voxel_size_mm = (2.21, 2.21, 2.21)

# Philips DICOM scaling parameters
rescale_slope = 2.598119
rescale_intercept = 0.0

# input CASToR reconstructed image
input_image = "/home/administrator/Secretária/castor_analysis/distancedriven_it3_aligned.nii"

# output calibrated image
output_image = "/home/administrator/Secretária/castor_analysis/CASTOR_FINAL_tests/pet_activity_Bq_per_mL_slopesPhilips.nii"


# -------------------------------------------------
# LOAD IMAGE
# -------------------------------------------------

nii = nib.load(input_image)
img = nii.get_fdata()

print("\nOriginal CASToR image stats:")
print("Min:", np.min(img))
print("Max:", np.max(img))


# -------------------------------------------------
# IF IMAGE IS COUNTS → CONVERT TO CPS
# -------------------------------------------------

img_cps = img 


# -------------------------------------------------
# COMPUTE VOXEL VOLUME
# -------------------------------------------------

voxel_volume_mm3 = voxel_size_mm[0] * voxel_size_mm[1] * voxel_size_mm[2]
voxel_volume_mL = voxel_volume_mm3 / 1000.0

print("\nVoxel volume (mL):", voxel_volume_mL)


# -------------------------------------------------
# COUNTS → ACTIVITY CONCENTRATION
# -------------------------------------------------

calibration_factor = (1e6) / (sensitivity * voxel_volume_mL)

print("Calibration factor:", calibration_factor, "Bq/mL per cps")

activity_image = img_cps * calibration_factor


# -------------------------------------------------
# APPLY PHILIPS DICOM RESCALE
# -------------------------------------------------

activity_image = activity_image * rescale_slope + rescale_intercept


# -------------------------------------------------
# PRINT FINAL STATISTICS
# -------------------------------------------------

print("\nFinal calibrated image stats:")
print("Min voxel value (Bq/mL):", np.min(activity_image))
print("Max voxel value (Bq/mL):", np.max(activity_image))


# -------------------------------------------------
# SAVE CALIBRATED IMAGE
# -------------------------------------------------

activity_nii = nib.Nifti1Image(activity_image, nii.affine, nii.header)
nib.save(activity_nii, output_image)

print("\nSaved calibrated image:")
print(output_image)
```
